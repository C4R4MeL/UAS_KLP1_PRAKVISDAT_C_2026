- **Indikator Bisnis & Risiko** (7 kolom): pertumbuhan revenue, sisa runway, product-market fit, turnover karyawan, probabilitas shutdown, risiko shutdown, flag kegagalan startup

## B. Data Cleaning
### Periksa Missing Values


```python
# Cek missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage (%)': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0]

if len(missing_df) == 0:
    print("✅ Tidak ada missing values pada dataset.")
else:
    print("⚠️ Kolom dengan missing values:")
    print(missing_df)
```

### Periksa Data Duplikat


```python
# Cek duplikat
duplicates = df.duplicated().sum()
print(f"Jumlah baris duplikat: {duplicates:,}")
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Setelah penghapusan duplikat: {df.shape[0]:,} baris")
else:
    print("✅ Tidak ada data duplikat.")
```

### Periksa Inkonsistensi Data Kategorikal


```python
# Periksa nilai unik setiap kolom kategorikal (dari dataset asli)
cat_cols = ['founder_type', 'economic_climate', 'industry', 'funding_stage',
            'work_mode', 'burnout_level', 'shutdown_risk', 
            'seeks_mental_health_support']

print("=" * 60)
print("NILAI UNIK - VARIABEL KATEGORIKAL")
print("=" * 60)
for col in cat_cols:
    vals = sorted(df[col].unique())
    print(f"\n{col} ({len(vals)} kategori):")
    for v in vals:
        count = (df[col] == v).sum()
        print(f"  • {v}: {count:,} ({count/len(df)*100:.1f}%)")
```

### Identifikasi Outlier (Metode IQR)


```python
# Deteksi outlier menggunakan metode IQR
num_cols = ['weekly_work_hours', 'sleep_hours', 'stress_score', 
            'decision_fatigue_score', 'investor_pressure_score',
            'cofounder_conflict_score', 'work_life_balance_score', 'burnout_score']

print("=" * 70)
print("LAPORAN OUTLIER (Metode IQR)")
print("=" * 70)

outlier_summary = []
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_summary.append({
        'Variabel': col,
        'Q1': round(Q1, 2),
        'Q3': round(Q3, 2),
        'IQR': round(IQR, 2),
        'Batas Bawah': round(lower, 2),
        'Batas Atas': round(upper, 2),
        'Jumlah Outlier': len(outliers),
        'Persentase (%)': round(len(outliers) / len(df) * 100, 2)
    })

outlier_df = pd.DataFrame(outlier_summary)
outlier_df
```

**Catatan:** Outlier tidak dihapus karena skor-skor ini berada pada skala tetap (1-10) dan outlier menurut IQR masih merupakan data yang valid secara kontekstual. Menghapusnya justru akan mengurangi representativitas dataset.

## C. Feature Engineering

Menambahkan kolom turunan baru dari data yang sudah ada untuk memperkaya analisis:

```python
# === Feature Engineering: Membuat kolom-kolom baru ===

# 1. funding_level: Kategori level pendanaan berdasarkan Funding_Stage
funding_map = {
    'Bootstrapped': 'Low', 'Pre-Seed': 'Low',
    'Seed': 'Medium', 'Series A': 'Medium',
    'Series B': 'High', 'Series C': 'High'
}
df['funding_level'] = df['funding_stage'].map(funding_map)

# 2. startup_size_category: Kategori ukuran startup berdasarkan Team_Size
def categorize_size(size):
    if size <= 10:
        return 'Small'
    elif size <= 50:
        return 'Medium'
    else:
        return 'Large'

df['startup_size_category'] = df['team_size'].apply(categorize_size)

# 3. funding_amount: Estimasi nominal pendanaan berdasarkan Funding_Stage
import numpy as np
amount_ranges = {
    'Bootstrapped': (5_000, 50_000),
    'Pre-Seed': (50_000, 250_000),
    'Seed': (250_000, 2_000_000),
    'Series A': (2_000_000, 15_000_000),
    'Series B': (15_000_000, 50_000_000),
    'Series C': (50_000_000, 100_000_000)
}
np.random.seed(42)
df['funding_amount'] = df['funding_stage'].apply(
    lambda x: np.random.randint(*amount_ranges.get(x, (0, 1)))
)

# Tampilkan hasil
print(f"Dataset setelah feature engineering: {df.shape[0]:,} baris × {df.shape[1]} kolom")
print(f"Kolom baru: funding_level, startup_size_category, funding_amount")
print()
print("Distribusi Funding Level:")
print(df['funding_level'].value_counts().to_string())
print()
print("Distribusi Startup Size Category:")
print(df['startup_size_category'].value_counts().to_string())
print()
print(f"Funding Amount range: ${df['funding_amount'].min():,.0f} - ${df['funding_amount'].max():,.0f}")
```
